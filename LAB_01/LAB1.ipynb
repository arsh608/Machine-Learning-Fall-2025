# TASK 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer_patient_dataset.csv")
df.head()
```

```python
print("Unique categories in target:", df['Level'].unique())
```

```python
print("Columns:", df.columns)
print("First 50 records:\n", df.head(50))
print("Last 30 records:\n", df.tail(30))
print("Info:")
print(df.info())
print("Shape:", df.shape)
print("Memory usage:", df.memory_usage().sum())
```

```python
print("Number of columns:", df.shape[1])
print("Number of samples:", df.shape[0])
```

```python
print(df['Level'].value_counts())

df['Level'].value_counts().plot(kind='bar', title="Category Distribution")
plt.show()
```

```python
G = df.loc[550:900, 'Gender':'Coughing of Blood']
G.to_csv("G2.csv", index=False)

print(G.describe(include='all'))
```

```python
H = df.iloc[-20:, df.columns.get_loc('Shortness of Breath'):df.columns.get_loc('Frequent Cold')+1]
H.to_csv("H2.csv", index=False)

print(H.describe(include='all'))
```

```python
I = df.loc[[110,220,360,440,656,778,202,889], :]
print("Average Alcohol Use:", I['Alcohol use'].mean())
```

```python
df['Age'].hist()
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()

```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.show()

sns.histplot(df['Frequent Cold'], kde=True)
plt.title("Distribution of Frequent Cold")
plt.show()
```

```python
sns.boxplot(x=df['Obesity'])
plt.title("Outliers in Obesity")
plt.show()
```
